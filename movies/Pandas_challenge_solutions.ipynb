```python
import pickle

with open('movie_data_dump.pkl','rb') as infile:
    movie_data = pickle.load(infile)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
mv_df = pd.DataFrame(movie_data)
mv_df.columns = pd.Index([u'director', u'domestic_total_gross', u'movie_title', u'rating', u'release_date', u'runtime'])
mv_df.head()
print mv_df.dtypes
```

```python
# Challenge 1: plot domestic total gross over time
fig = plt.figure(figsize=(15,5))
mv_df = mv_df.sort(columns=['release_date'])
mv_df.plot(x='release_date',y='domestic_total_gross')
```

```python
# Challenge 1: plot domestic total gross over time
fig=plt.figure(figsize=(15,5))
plt.plot(mv_df['release_date'],mv_df['domestic_total_gross'],'gx')
```

```python
# Challenge 2: plot runtime vs domestic total gross

# solution 1 -- sort dataframe by runtime and plot a line
mv_df_runtimesort = mv_df.sort(columns=['runtime'])
# this looks really weird if you don't group it because there is more than one y per x
runtime_means = mv_df_runtimesort.groupby('runtime').aggregate('mean')

ax = runtime_means.plot()

# I wanted to make it bigger and it wouldn't respond to the trick I used elsewhere, so I got in there and made it bigger. 
fig = ax.get_figure()
fig.set_size_inches(20,5)

# then when it was bigger I wanted more ticks on the x axis. 
tx = ax.xaxis.set_ticks(range(50,260,10))
```

```python
# Challenge 2: plot runtime vs domestic total gross

# solution 2 - don't worry about sorting if you use a marker for each datapoint
plt.figure(figsize=(15,5))
mv_df.plot(x='runtime',y='domestic_total_gross',style='kx')
```

```python
# Challenge 3: group your data by Rating and find the average runtime and domestic total gross at each level of Rating
pd.options.display.float_format='{:.1f}'.format
df_by_rating = mv_df.groupby('rating')['runtime','domestic_total_gross'].mean()
df_by_rating
```

```python
# Challenge 3 bonus plot
axs =df_by_rating.plot(subplots=True)
axs[0].get_figure().set_size_inches(13,8)
```

```python
# Challenge 4: make one figure with (N=the number of MPAA ratings there are) subplots, and in each plot the release date 
#              vs the domestic total gross

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 8))

gbr = mv_df.groupby('rating')
for i, (name, group) in enumerate(gbr):
    group.plot(x='release_date',y='domestic_total_gross',ax=axes[int(i>2)][i%3],title=name)

# these are ugly and I would change them but I am tired so I'm stopping for now :) 
```

```python
# Challenge 5: what director in your dataset has the highest gross per movie?

grouped_by_director = mv_df.groupby('director')
director_means = grouped_by_director['domestic_total_gross'].agg([np.count_nonzero,np.mean]).sort('mean',ascending=False)
director_means.head(20)
```

```python
# Challenge 5: what director in your dataset has the highest gross per movie?
#Here is the same table with the one-hit-wonders removed. 

dir_means_multiple_movies = director_means[director_means['count_nonzero']>1]
dir_means_multiple_movies.sort(columns=['mean'],ascending=False).head(15)
```

```python
# Challenge 6: bin your dataset into years (if applicable) and make a bar graph with error bars of gross each year

ygrouped = mv_df.groupby(mv_df['release_date'].map(lambda x: x.year))
yearmeans = ygrouped['domestic_total_gross'].agg({'mean':'mean','sem':lambda x: np.std(x)/len(x)})
fig = plt.figure(figsize=(15,10))
axs = fig.add_axes()
p = yearmeans.plot(y='mean',kind='bar',yerr=yearmeans['sem'],ax=axs,title='Average gross per film each year')
```
