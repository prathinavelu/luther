```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import seaborn as sns
import re
from datetime import date
from collections import defaultdict

%matplotlib inline

with open('all_movie_data_combined_new.pkl','rb') as infile:
    movie_data_all = pickle.load(infile)

movie_data_all = [x for x in movie_data_all if x != None]
```

```python
movie_data_clean = [x for x in movie_data_all if x['gross'] != None and
                                             x['release date'] != None and
                                             len(x['release date'].split()) == 3 and
                                             x['remake'] != None
                                             ]

print len(movie_data_clean)
```

```python
for m in movie_data_clean:
    m['year'] = int(m['release date'][-4:])
    m['month'] = ''.join(re.findall("[a-zA-Z]+", m['release date']))
    try:
        m['remake']['year'] = int(m['remake']['release date'][-4:]) 
        m['remake']['month'] = ''.join(re.findall("[a-zA-Z]+", m['remake']['release date']))
    except:
        pass
    try:
        m['gross'] = int(m['gross'])
    except:
        m['gross'] = None
    m['remake'] = defaultdict(str,m['remake'])

#m_v_clean = [x for x in movie_data_clean if x['gross'] != None]
k = [x for x in movie_data_clean[0].keys() if x != 'remake']

for i in k:
    s = 'original_'+i
    for m in movie_data_clean:
        m[s] = m['remake'][i]
```

```python
for m in movie_data_clean:
    if m['original_color'] != m['color']:
        m['colorchange'] = 1
    else:
        m['colorchange'] = 0
    try:    
        m['gross'] = int(m['gross'])
    except:
        m['gross'] = None
    try:    
        m['original_gross'] = int(m['original_gross'])
    except:
        m['original_gross'] = None
```

```python
languages = []
for m in movie_data_clean:
    if m['original_language'] not in languages:
        languages.append(m['original_language'])
        
#print languages

for m in movie_data_clean:
    m['american_origin'] = 0
    m['european_origin'] = 0
    m['asian_origin'] = 0
    if m['original_language'] == 'English':
        m['american_origin'] = 1
    elif (m['original_language'] == 'French' or m['original_language'] == 'Swedish' or m['original_language'] == 'Norwegian' or m['original_language'] == 'Spanish' or  m['original_language'] == 'Danish' or m['original_language'] == 'German' or m['original_language'] == 'Italian' or m['original_language'] == 'Dutch'):
        m['european_origin'] = 1
    elif m['original_language'] == None:
        pass
    else:
        m['asian_origin'] = 1
    if m['gross'] > 1000000000:
        m['gross'] = int(str(m['gross'])[:-4])
    if m['original_gross'] > 1000000000:
        m['original_gross'] = int(str(m['original_gross'])[:-4])
```

```python
movies = [x for x in movie_data_clean if x['gross'] != None and x['original_gross'] != None ]
```

```python
import math
from inflation_cpi import get_cpi_values

cpi = get_cpi_values()
c= cpi[2014]

for m in movies:
    m['i_gross'] = m['gross']*(1 + (c - cpi[m['year']])/cpi[m['year']])
    m['i_original_gross'] = m['original_gross']*(1 + (c - cpi[m['original_year']])/cpi[m['original_year']])
    m['log_i_og_gross'] = math.log(m['i_original_gross'])
    m['log_i_gross'] = math.log(m['i_gross'])
    m['yearsbetween'] = m['year'] - m['original_year']
```

```python
import itertools
import statsmodels.formula.api as sm

def findsubsets(S,m):
    lst = list(itertools.combinations(S, m))
    newlist = []
    for l in lst:
        newlist.append(list(l))
    return newlist

l = ['yearsbetween','american_origin','european_origin','asian_origin','original_gross']


all = []
for i in range(1,len(l)+1):
    all = all + findsubsets(l,i)
    

results = []
for a in all:
    b = a + ['Ones']
    X = df[b]
    Y = df['gross']
    model = sm.OLS(Y,X).fit()
    results.append([a,model.rsquared])

for r in results:
    print r
```

```python
l = int(len(df)*.75)
factors = []
for i in range(2,8):
    y = 'yearsbetween**'+str(i)
    df[y] = df['yearsbetween'].apply(lambda x: x**i)
    factors.append(y)


movies_sample = df.reindex(np.random.permutation(df.index))
movies_training = movies_sample[:l]
movies_test = movies_sample[l:]

X = movies_training[['yearsbetween','original_gross', 'Ones']+factors]
Y = movies_training['gross']
model = sm.OLS(Y,X).fit()
model.summary()
```

```python
import math
from sklearn import metrics

X_test = movies_test[['yearsbetween','original_gross', 'Ones']+factors]
actual = movies_test['gross']
predicted = model.predict(X_test)

mse = math.sqrt(metrics.mean_squared_error(predicted,actual))
print mse*2
```

```python
from statsmodels.sandbox.regression.predstd import wls_prediction_std

f = plt.figure(figsize=[15,7.5])
plt.plot(movies_test['yearsbetween'],actual,'bo')
plt.plot(movies_test['yearsbetween'],predicted,'gd')
prstd, conf_lower, conf_upper = wls_prediction_std(model)
plt.plot(movies_training['yearsbetween'], conf_lower, 'yo')
plt.plot(movies_training['yearsbetween'], conf_upper, 'yo')
```
